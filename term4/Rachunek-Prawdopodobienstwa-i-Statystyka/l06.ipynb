# RPiS
2D and 3D linear regression

```julia
using CSV
using Plots
```

```julia
data = CSV.read("klimat.csv", delim = ";", types = [String, Float64, Float64, Int, Float64, Int])
```

## 10.

```julia
temperature = data[5]
height = data[6]

Y = temperature
X = hcat(ones(Y), height)

coeffs = X'*X \ X'*Y
@show coeffs

plot(height, temperature, seriestype = :scatter, leg = false)
h = [minimum(height), maximum(height)]
plot!(h, [dot(coeffs, [1, h]) for h in h])
```

## 11.

```julia
temperature = data[5]
length = data[3]
width = data[2]

Y = temperature
X = hcat(ones(Y), length, width)

coeffs = X'*X \ X'*Y
@show coeffs

plot(length, width, temperature, seriestype = :scatter, leg = false)
l = [minimum(length), maximum(length)]
w = [minimum(width), maximum(width)]
plot!(l, w, [dot(coeffs, [1, l, w]) for (l, w) in zip(l, w)])
```
