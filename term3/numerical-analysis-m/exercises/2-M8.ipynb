Using matrix-based least squares linear regression.

Problem: for $Y=aX+b$ find parameters $a$, $b$, where $X$ and $Y$ are given as vectors.

```julia
X = [0 10 20 30 40 80 90 95]
Y = [68.0 67.1 66.4 65.6 64.6 61.8 61.0 60.0]

X_ = vcat(X, ones(1, length(X)))
```

```julia
# Left division operator: A \ B = inv(A) * B

a, b = X_ * X_' \ X_ * Y'

@show a
@show b
```

```julia
ends = [X[1], X[end]]
scatter(X, Y, color="b", marker="x")
plot(ends, a*ends + b, linewidth=0.3, color="k");
```
