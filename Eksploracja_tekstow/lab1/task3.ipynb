```python
import unicodedata

import nltk
nltk.download('punkt');
```

```python
def strip_punctuation(word):
    if not word:
        return word
    lw = len(word)
    li, ri = 0, lw - 1
    ltokens, rtokens = [], []
    while li < lw and unicodedata.category(tok := word[li]).startswith('P'):
        li += 1
        ltokens.append(tok)
    while ri > 0 and unicodedata.category(tok := word[ri]).startswith('P'):
        ri -= 1
        rtokens.append(tok)
    return *ltokens, word[li : ri + 1], *reversed(rtokens)


def my_tokenize(text):
    return [
        token
        for word in text.split()
        for token in strip_punctuation(word)
        if token
    ]
```

```python
PRINT_MATCHING = True


def display_tokens(tokens):
    return '  '.join(tokens)


def print_diff(my_tokens, nltk_tokens):
    if my_tokens == nltk_tokens:
        if PRINT_MATCHING:
            print(
                'MATCH',
                display_tokens(my_tokens),
                sep='\n',
                end='\n\n',
            )
        return
    print(
        '<<<<<MY',
        display_tokens(my_tokens),
        '-----',
        display_tokens(nltk_tokens),
        'NLTK>>>>>',
        sep='\n',
        end='\n\n',
    )


with open('data/cytaty.txt') as f:
    for line, _ in zip(f, range(50)):
        print_diff(
            my_tokenize(line),
            nltk.tokenize.word_tokenize(line),
        )
```
