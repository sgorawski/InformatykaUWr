```python
%matplotlib inline

import matplotlib.pyplot as plt

import scores
```

```python
import spectral
img = spectral.open_image('/pio/scratch/2/ED2015/APEX/APEX_OSD_Package_1.0/APEX_OSD_Package_1.0/APEX_OSD_V1_calibr_cube.hdr')
data = img.load()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(init='k-means++', n_clusters=4, n_init=1)

data_2d = data.reshape(-1, 285)
kmeans.fit(data_2d)
```

```python
def show_apex(alg):
    _, ax = plt.subplots(1, 1, figsize=(10, 15))
    im = alg.labels_.reshape(data.shape[:2])
    ax.axis('off')
    ax.imshow(im)
```

```python
show_apex(kmeans)
```

```python
kmeans2 = KMeans(init='k-means++', n_clusters=8, n_init=1)
kmeans2.fit(data_2d)
```

```python
show_apex(kmeans2)
```

```python
kmeans3 = KMeans(init='k-means++', n_clusters=2, n_init=1)
kmeans3.fit(data_2d)
```

```python
show_apex(kmeans3)
```

```python
kmeans4 = KMeans(init='k-means++', n_clusters=12, n_init=1)
kmeans4.fit(data_2d)
```

```python
show_apex(kmeans4)
```
