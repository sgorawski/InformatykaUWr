Testing iterative method of computing the area of a regular polygon, based on the formula:

$$P_n=\frac{1}{2}n\sin\frac{2\pi}{n}$$

```julia
t = 128 #precision

setprecision(t) do
    # Initial values
    s = BigFloat(1.0)
    c = BigFloat(0.0)
    P = BigFloat(2.0)
    
    for k = 3:2t
        s = sqrt((1-c) / 2)
        c = sqrt((1+c) / 2)
        P = 2^(k-1) * s
        @show P
    end
end
```

# M3.7
---
Finding function $f$ for which Newton method is of linear convergence. Result: $f=x^2$.

```julia
t = 128 #precision
f(x) = x^2
df(x) = 2x

setprecision(t) do
    x = BigFloat(10.0)
    while abs(x^2 - 0) > 2.0^-t
        x -= f(x) / df(x)
        @show x
    end
end
```

# M3.8
---
Implementation of the Steffensen method on the example of equation $e^{-x} - \sin{x} = 0$ + convergence tests. Result: cubic convergence.

```julia
# Steffensen method h_n
hn(f, c) = f(c)^2 / (f(c+f(c)) - f(c));
```

```julia
t = 128
f(x) = exp(-x) - sin(x)

setprecision(t) do
    root = BigFloat(9.42469725473852)
    c = BigFloat(10.0)

    while abs(f(c)-0) > 2.0^-t
        c -= hn(f, c)
        ϵ = c - root
        @show ϵ
    end
end
```
