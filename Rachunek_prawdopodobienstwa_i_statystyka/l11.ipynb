## 3.

```julia
X = [1, 3, 4, 6, 8, 9, 11, 14]
Y = [1, 2, 4, 4, 5, 7, 8, 9]

X_ = hcat(ones(X), X)
coeffs = X_'*X_ \ X_'*Y
@show coeffs

plot(X, Y, seriestype = :scatter, leg = false)
x = [minimum(X), maximum(X)]
plot!(x, [dot(coeffs, [1, x]) for x in x])
```

## 4.

$$PV^k = C \qquad k, C - const$$
$$\log P + k\log V = \log C$$
$$-\log C + k\log V = -\log P$$

```julia
V = [54.3, 61.8, 72.4, 88.7, 118.6, 194.0]
P = [61.2, 49.5, 37.6, 28.4, 19.2, 10.1]

X = hcat(ones(V), log.(V))
Y = -log.(P)
minus_log_C, k = X'*X \ X'*Y
@show minus_log_C
@show k

C = exp(-minus_log_C)
@show C

p = C / 100^k  # for V = 100
@show p;
```

## 5.

```julia
X = [1.2, 1.8, 3.1, 4.9, 5.7, 7.1, 8.6, 9.8]
Y = [4.5, 5.9, 7.0, 7.8, 7.2, 6.8, 4.5, 2.7]

X_ = hcat(ones(X), X, X.^2)
coeffs = X_'*X_ \ X_'*Y
@show coeffs

plot(X, Y, seriestype = :scatter, leg = false)
x = linspace(minimum(X), maximum(X), 100)
plot!(x, [dot(coeffs, [1, x, x^2]) for x in x])
```

```julia
Z = [64, 71, 53, 67, 55, 58, 77, 57, 56, 51, 76, 68]
X = [57, 59, 49, 62, 51, 50, 55, 48, 52, 42, 61, 57]
Y = [8, 10, 6, 11, 8, 7, 10, 9, 10, 6, 12, 9]

X_ = hcat(ones(Z), X, Y)
coeffs = X_'*X_ \ X_'*Z
@show coeffs

plot(X, Y, Z, seriestype = :scatter, leg = false)
x = [minimum(X), maximum(X)]
y = [minimum(Y), maximum(Y)]
plot!(x, y, [dot(coeffs, [1, x, y]) for (x, y) in zip(x, y)])
```
